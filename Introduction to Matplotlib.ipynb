# Which ploting method to use ? (Pyplot vs Matplotlib OO Method)
* To plot something quick use pyplot
* To plot something more advanced use Matplotlib OO Method


```python
heart_disease.head()
```

```python
age_over_50 = heart_disease[heart_disease["age"] > 50]
age_over_50.head()
```

```python
# Pyplot Method
age_over_50.plot(x="age",
                 y="chol",
                 kind="scatter",
                 figsize=(10,5),
                 c="target");
```

```python
# Matplotlib OO Method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,5))
age_over_50.plot(kind="scatter",
                 x="age",
                 y="chol",
                 c="target",
                 ax=ax)
ax.set_xlim([45,100])
ax.set_ylim([50,700])
```

```python
plt.style.use("seaborn-notebook")
# OO Method from scratch 
fig,ax = plt.subplots(figsize=(12,8))

# Plot the data
scatter = ax.scatter(x=age_over_50["age"],
                     y=age_over_50["chol"],
                     c=age_over_50["target"])

# Coustomise the plot
ax.set(title="Heart Disease and Cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol")
ax.set_ylim([0,700]) 
ax.set_xlim([50,80])

# Add a Legend 
ax.legend(*scatter.legend_elements(),title="Target")

# Add a Horizontal line
ax.axhline(age_over_50["chol"].mean(),linestyle="--");
```

```python
# Create a subplo for Heart disease data using OO Method
#plt.style.use("tableau-colorblind10")
# Creat a fig and axis
fig, (ax0,ax1) = plt.subplots(ncols= 1, nrows=2, figsize = (12,10),sharex=True)

# Add data ax0
scatter =ax0.scatter(x=age_over_50["age"],
                     y=age_over_50["chol"],
                     c=age_over_50["target"])

# Customize ax0
ax0.set(title= "Heart Disease and Cholesterol Levels",
        ylabel="Cholestarol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=age_over_50["chol"].mean(),linestyle ="--")

# Add data to ax1
scatter = ax1.scatter(x=age_over_50["age"],
                      y=age_over_50["thalach"],
                      c=age_over_50["target"])

# Customise ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a Legend to AX1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax1.axhline(y=age_over_50["thalach"].mean(),linestyle="--")

# Add a titile to the figure
fig.suptitle("Heart Disease Data Analysis",fontsize=16 ,fontweight="bold");
```

```python
# Create a subplo for Heart disease data using OO Method
plt.style.use("seaborn-whitegrid")
# Creat a fig and axis
fig, (ax0,ax1) = plt.subplots(ncols= 1, nrows=2, figsize = (12,10),sharex=True)

# Add data ax0
scatter =ax0.scatter(x=age_over_50["age"],
                     y=age_over_50["chol"],
                     c=age_over_50["target"],
                     cmap="winter")           # This changes the colour scheme

# Customize ax0
ax0.set(title= "Heart Disease and Cholesterol Levels",
        ylabel="Cholestarol")

# Change the axis limit

ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=age_over_50["chol"].mean(),linestyle ="--")

# Add data to ax1
scatter = ax1.scatter(x=age_over_50["age"],
                      y=age_over_50["thalach"],
                      c=age_over_50["target"],
                      cmap="winter")

# Customise ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a Legend to AX1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax1.axhline(y=age_over_50["thalach"].mean(),linestyle="--")

# Add a titile to the figure
fig.suptitle("Heart Disease Data Analysis",fontsize=16 ,fontweight="bold");
```
